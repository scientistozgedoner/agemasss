```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
from matplotlib.ticker import NullFormatter, FixedLocator
import numpy as np
cluster = pd.read_csv('/kaggle/input/logagemass/logage_mass.csv')
massc= cluster.iloc[:,0].values
logagec= cluster.iloc[:,1].values
Mv=pd.read_csv('/kaggle/input/logagemass/Mv_limit.csv')
massl= Mv.iloc[:,0].values
logagel= Mv.iloc[:,1].values

fig, ax= plt.subplots(figsize=(8,9))

ax.plot(logagec,massc,color="black" , ls='None' , marker="o", markersize="4")
ax.plot(logagel,massl,color="red")

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
ax.tick_params(width=3)
ax.legend()
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', labelsize=12)

fontsize = 16

for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')

import sympy as sp
ab = sp.Symbol("tau")



ax.set_xlabel("Log(\yr)", fontsize=16, fontweight='bold')
ax.set_ylabel("Log($M/M_{\odot}$)",fontsize=17, fontweight='bold')
ax.set_xlim(6,9.5)
ax.set_ylim(2.2,6)
ax.xaxis.set_major_locator(FixedLocator(np.arange(6,11,0.5)))

ax.legend()

plt.savefig('mass-age.png',dpi=300, bbox_inches='tight')
plt.show()

```
